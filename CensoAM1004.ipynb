#Visualização do Dados

```python
sns.countplot(x = base_censo['income']);
```

```python
plt.hist(x = base_censo['age'], edgecolor = 'black');
plt.subplots_adjust(wspace = 0.4)
plt.show()
```

```python
plt.hist(x = base_censo['education-num'], edgecolor = 'black');
```

```python
plt.hist(x = base_censo['hour-per-week'], edgecolor = 'black');
```

```python
profissao_idade = px.treemap(base_censo, path=['workclass', 'age'])
profissao_idade.show()
```

```python
ocupacao_relacao_idade = px.treemap(base_censo, path=['occupation', 'relationship', 'age'])
ocupacao_relacao_idade.show()
```

```python
cordas_ocupacao_relacao = px.parallel_categories(base_censo, dimensions=['occupation', 'relationship'])
cordas_ocupacao_relacao.show()
```

```python
cordas_profissao_area_renda = px.parallel_categories(base_censo, dimensions=['workclass', 'occupation', 'income'])
cordas_profissao_area_renda.show()
```

```python
cordas_educacao_renda = px.parallel_categories(base_censo, dimensions=['education', 'income'])
cordas_educacao_renda.show()
```
